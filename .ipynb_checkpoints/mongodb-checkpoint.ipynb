```python
import pymongo
```

```python
from pprint import pprint
```

```python
client = pymongo.MongoClient('localhost',27017)
```

```python
db = client.imdb
db
```

```python
film_collection = db.films
serie_collection = db.series
```

```python
#Compter le nombre de collection
len(db.list_collection_names())
```

```python
#Compter le nombre de document dans une collection
films = film_collection.count_documents({}) #[1:] car collection avec un objet ajouté pour la création de la database
series = serie_collection.count_documents({})

films, series
```

```python
#Compter le nombre de films sans titre-original
films = film_collection.count_documents({'original_title':None})
films
```

```python
#Films non parues aux USA et sortis aprés 1994
films = film_collection.count_documents({"$and":[{"date": {"$lt": 1994}},{"country": {"$ne": "USA"}}]})
films
```

```python
#Tous les films
films = [film for film in film_collection.find({})]
films
```

```python
#Toutes les séries
series = [serie for serie in serie_collection.find({})]
series
```

```python
#Tous les films mais seulement les champs titre, note et description
films = [film for film in film_collection.find({}, {"score":1, "title":1, "description":1})]
films
```

```python
#Tous les films mais seulement les champs titre, note et description sans le champ id
films = [film for film in film_collection.find({}, {"score":1, "title":1, "description":1, "_id":0})]
films
```

```python
#Toutes les séries dont la note est inférieur à 9
series = [serie for serie in serie_collection.find({"score":{"$lt":9}})]
series
```

```python
#Tous les titres de film triè dans l'ordre alphabétique
films = [film for film in film_collection.find({}, {"_id":0,"title":1}).sort("title")]
films
```

```python
#Titre et note du top 5 selon la note des films
films = [film for film in film_collection.find({},{"_id": 0, "title": 1, "score": 1}).sort("score",-1)][:5]
films
```

```python
films = [film for film in film_collection.find({},{"_id": 0, "title": 1, "score": 1}).sort("score",-1)][5:10]
films
```

```python
#Toutes les séries dont la note est supérieure à 9
series = [serie for serie in serie_collection.find({"score":{"$gt":9}})]
series
```

```python
#Titres de film avec note inférieur à 8.5 et date > 2002
films = [film for film in film_collection.find({"$and":[{"score": {"$lt": 8.5}},{"date": {"$gt": 2002}}]}, {"_id": 0,"title":1})]
films
```

```python
#Titres et notes des series avec note inférieure à 9.4 mais supérieure à 9
series = [serie for serie in serie_collection.find({"$and":[{"score": {"$lt": 9.4}}, {"score": {"$gt": 9}}]}, {"_id": 0, "title": 1, "score": 1})]
series
```

```python
#Titres, notes et descriptions des series avec note inférieure à 9.5 mais supérieure à 9
series = [serie for serie in serie_collection.find({"$and":[{"score": {"$lt": 9.5}}, {"score": {"$gt": 9}}]}, {"_id": 0, "title": 1, "score": 1, "description": 1})]
series
```

```python
# Titres de film dont année supérieure à 2000 ou note supérieure à 9
films = [film for film in film_collection.find({"$or":[{"score": {"$gt": 9}}, {"date": {"$gt": 2000}}]}, {"_id": 0, "title": 1})]
films
```

```python
#Titres de films avec note supérieure à 9 et durée < 2h45
films = [film for film in film_collection.find({"$and":[{"score": {"$gt": 9}}, {"time": {"$lt": "2h 45m"}}]}, {"_id": 0, "title": 1})]
films
```

```python
#Titres de séries contenant au moins un chiffre
series = [serie for serie in serie_collection.find({"title": {"$regex":"\d+"}}, {"_id": 0, "title": 1})]
series
```

```python
#Titres de films contenant "the" majuscule ou non
films = [film for film in film_collection.find({"title": {"$regex":"the"}}, {"_id": 0, "title": 1})]
films
```
